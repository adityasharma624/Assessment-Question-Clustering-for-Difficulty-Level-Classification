```python
import pandas as pd
```

```python
df = pd.read_csv("../datasets/leetcode_dataset.csv")
print(df.columns)
```

```python
df = df[['title', 'description', 'related_topics']]
```

```python
df = df.apply(lambda x: x.astype(str).str.lower() if x.dtype == 'object' else x)
df = df.apply(lambda col: col.str.replace(r"\s+", " ", regex=True).str.strip() if col.dtype == "object" else col)
```

```python
df['cleaned_text'] = df['title'] + " " + df['description'] + " topics: " + df['related_topics']
df = df[['cleaned_text']]
df.head()
```

```python
output_path = "../datasets/cleaned_leetcode_dataset.csv"
df.to_csv(output_path, index=False)
print(f"Cleaned dataset saved to {output_path}")
```
